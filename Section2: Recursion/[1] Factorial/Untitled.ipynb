```python
def fact(n):
    if n == 0:
        return 1
    
    return n * fact(n-1)


result = int(input("Input n: "))
res = fact(result)

print("Result: ",res)
```

```python
#improve solution

def fact(n):
    result = 1
    while n != 0:
        result *= n
        n -= 1
    return result
```
